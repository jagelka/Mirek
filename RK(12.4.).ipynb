```python
from __future__ import division
import matplotlib.pyplot as plt

def fx(x,y,z):
    a = 10.0
    return a*(x-y)

def fy(x,y,z):
    b = 28.0
    return x*(b-z)-y

def fz(x,y,z):
    c = 8.0/3
    return x*y-c*z

def RK4x(f,x,y,z,h):
    K0 = h*f(x,y,z)
    K1 = h*f(x+K0/2,y,z)
    K2 = h*f(x+K1/2,y,z)
    K3 = h*f(x+K1,y,z)
    return x+(K0+2*K1+2*K2+K3)/6

def RK4y(f,x,y,z,h):
    K0 = h*f(x,y,z)
    K1 = h*f(x,y+K0/2,z)
    K2 = h*f(x,y+K1/2,z)
    K3 = h*f(x,y+K2,z)
    return y+(K0+2*K1+2*K2+K3)/6

def RK4z(f,x,y,z,h):
    K0 = h*f(x,y,z)
    K1 = h*f(x,y,z+K0/2)
    K2 = h*f(x,y,z+K1/2)
    K3 = h*f(x,y,z+K2)
    return z+(K0+2*K1+2*K2+K3)/6


x = [1.0]
y = [1.0]
z = [1.0]
for i in range(1000):
    x.append(RK4x(fx,x[-1],y[-1],z[-1],0.001))
    y.append(RK4y(fy,x[-2],y[-1],z[-1],0.001))
    z.append(RK4z(fz,x[-2],y[-2],z[-1],0.001))

plt.plot(x,z,'*')
plt.figure()
plt.plot(x,y,'*')
plt.figure()
plt.plot(y,z,'*')
```
